```python
f = input("Enter given example text file.txt4: ")
filename = open(f,"r")
def count_totalcharater(filename):
    try:
        with open(filename,encoding="utf8") as f_obj:
            contents = f_obj.read()
            Total_chracters = len(contents)
            print("The file " + filename + " has about " + str(Total_chracters) + " Total Characters.")
    except FileNotFoundError:
        msg = "Sorry, the file " + filename + " does not exist."
        print(msg)
        
def count_word(filename):
    try:
        with open(filename,encoding="utf8") as f_obj:
            contents = f_obj.read()
            word = contents.split()
            num_word = len(word)
            print("The file " + filename + " has about " + str(num_word) + " words.")
    except FileNotFoundError:
        msg = "Sorry, the file " + filename + " does not exist."
        print(msg)
        
def count_space(filename):
    try:
        with open(filename,encoding="utf8") as f_obj:
            contents = f_obj.read()
            space = contents.split(" ")
            num_space = len(space)
            print("The file " + filename + " has about " + str(num_space) + " spaces.")
    except FileNotFoundError:
        msg = "Sorry, the file " + filename + " does not exist."
        print(msg)
        
def count_lines(filename):
    try:
        with open(filename,encoding="utf8") as content_file:
            contents = content_file.read()
            lineCount = contents.split("\n")
            num_line = len(lineCount)
            print("The file " + filename + " has about " + str(num_line) + " Lines.")

    except FileNotFoundError:
        msg = "Sorry, the file " + filename + " does not exist."
        print(msg)

def count_tabs(filename):
    try:
        with open(filename,encoding="utf8") as content_file:
            contents = content_file.read()
            tabsCount = contents.split("\t")
            num_tab = len(tabsCount)
            print("The file " + filename + " has about " + str(num_tab) + " tabs.")

    except FileNotFoundError:
        msg = "Sorry, the file " + filename + " does not exist."
        print(msg)
def count_paragraph(filename):
    try:
        with open(filename,encoding="utf8") as f_obj:
            contents = f_obj.read()
            paragraph = contents.split('\n\t')
            num_paragraph = len(paragraph)
            print("The file " + filename + " has about " + str(num_paragraph) + " paragraphs.")
    except FileNotFoundError:
        msg = "Sorry, the file " + filename + " does not exist."
        print(msg)
        
def count_nonspacecharater(filename):
    try:
        with open(filename,encoding="utf8") as f_obj:
            contents = f_obj.read()
            Total_chracters = len(contents)
            word = contents.split()
            num_word = len(word)
            space = contents.split(" ")
            num_space = len(space)
            tabsCount = contents.split("\t")
            num_tab = len(tabsCount)
            lineCount = contents.split("\n")
            num_line = len(lineCount)
            paragraph = contents.split('\n\t')
            num_paragraph = len(paragraph)
            num_nonspacecharater = Total_chracters - (num_space + num_word + num_tab + num_line + num_paragraph)
            print("The file " + filename + " has about " + str(num_nonspacecharater) + " nonspacecharater.")
    except FileNotFoundError:
        msg = "Sorry, the file " + filename + " does not exist."
        print(msg)
        
filename = [f]

for filename in filename:
    count_totalcharater(filename)
    count_word(filename)
    count_space(filename)
    count_tabs(filename)
    count_lines(filename)
    count_paragraph(filename)
    count_nonspacecharater(filename)

```

```python
f = input("Enter given Html example file to parse: ")
filename = open(f,"r")        
def count_Tags(filename):
    try:
        with open(filename,encoding="utf8") as f_obj:
            fileContents = f_obj.read()
            print (fileContents)
            pairedTags = fileContents.count('</')
            unPairedTags = fileContents.count('/>')
            totalTags = pairedTags + unPairedTags
            print("The file " + filename + " has about " + str(totalTags) + " tags.")
    except FileNotFoundError:
        msg = "Sorry, the file " + filename + " does not exist."
        print(msg)
        
filename = [f] 
     
for filename in filename:
    count_Tags(filename)
```
